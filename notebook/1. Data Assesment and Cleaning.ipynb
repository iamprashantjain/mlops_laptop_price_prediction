```python
# Function to convert memory capacity to GB
def convert_to_gb(memory):
    if 'TB' in memory:
        # Extract the numeric part and convert TB to GB
        return float(memory.replace('TB', '').strip()) * 1024
    elif 'GB' in memory:
        # Extract the numeric part and keep GB
        return float(memory.replace('GB', '').strip())
    else:
        # Handle cases where the unit might be missing or unknown
        raise ValueError(f"Unknown memory unit in: {memory}")
```

```python
# Apply conversion function to the DataFrame
messy_df['memory_capacity'] = messy_df['memory_capacity'].apply(convert_to_gb)
```

```python
# Define the brand mapping
brand_mapping = {
    'NVIDIA': 'NVIDIA',
    'AMD': 'AMD',
    'Intel': 'Intel',
    'Qualcomm': 'Qualcomm',
    'Apple GPU': 'Apple GPU'
}

# Function to determine graphics card brand
def determine_brand(graphics_card):
    if pd.isna(graphics_card):
        return 'Others'
    for keyword, brand in brand_mapping.items():
        if keyword in graphics_card:
            return brand
    return 'Others'
```
