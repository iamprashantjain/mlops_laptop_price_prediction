Here’s a summary of the columns and potential redundancies:

- productName: Descriptive name of the product.
- price: The price of the product.
- rating: Product rating.
- specScore: Some sort of specification score.
- features: A concatenated string of features.
- brand: Brand of the product.
- processor: Processor details.
- storage: Storage capacity.
- cores: Number of processor cores.
- ram: Amount of RAM.
- graphics: Graphics card details.
- storage_type: Type of storage (e.g., SSD).
- ram_type: Type of RAM (e.g., DDR4).
- resolution: Resolution in 'width x height' format.
- graphics_category: Category of graphics (e.g., NVIDIA).
- resolution_category: Category of resolution (e.g., Quad HD or higher).
- width: Extracted width from resolution.
- height: Extracted height from resolution.
- aspect_ratio: Aspect ratio calculated from width and height.
- resolution_area: Area in pixels calculated from width and height.

#### remove duplicate columns 

- Remove features: This column is a concatenated string of various features and can be considered redundant if you have separate columns for those features.
- Remove resolution: You can derive width, height, aspect_ratio, and resolution_area from it.
- Remove resolution_category: This might be redundant if aspect_ratio and resolution_area are sufficiently informative for your model.
- Remove graphics_category: This might overlap with graphics.

```python
data_df_cleaned = df.drop(columns=['features', 'resolution', 'resolution_category', 'graphics_category','productName'])
```
