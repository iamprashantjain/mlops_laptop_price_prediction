```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_smartprix_laptop4.csv")
```

### Extract Width and Height

```python
df[['width', 'height']] = df['resolution'].str.split(' x ', expand=True).astype(int)
```

#### Create Additional Features

- Aspect Ratio: The ratio of width to height.
- Resolution Area: The total number of pixels (width × height).

```python
df['aspect_ratio'] = df['width'] / df['height']
df['resolution_area'] = df['width'] * df['height']
```
