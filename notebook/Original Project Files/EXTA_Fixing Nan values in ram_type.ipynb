```python
df['features'] = df['features'].str.replace("]","").str.replace("[","")
```

```python
df['features'].iloc[45].split(",")
```

```python
import re

def extract_ram_type(features_str):
    # Define regex patterns for different RAM types
    patterns = [
        r'\bDDR4\b',
        r'\bDDR5\b',
        r'\bLPDDR5X\b',
        r'\bLPDDR4X\b'
    ]
    
    # Search for patterns in the features string
    for pattern in patterns:
        match = re.search(pattern, features_str, re.IGNORECASE)
        if match:
            return match.group(0)
    
    # Return 'Unknown' if no pattern is matched
    return 'Unknown'
```

```python
df.loc[df['ram_type'].isna(), 'ram_type'] = df.loc[df['ram_type'].isna(), 'features'].apply(extract_ram_type)
```
